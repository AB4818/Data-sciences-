```python
train=walmart.head(147)
test=walmart.tail(12)
```

```python
train
```

```python
test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import statsmodels.formula.api as smf
```

Linear model

```python
linear_model =smf.ols('Footfalls~t',data=train).fit()
pred_linear = pd.Series(linear_model.predict(pd.DataFrame(test['t'])))
rmse_linear=np.sqrt(mean_squared_error(test['Footfalls'],pred_linear))
rmse_linear
```

Exponential model

```python
exp=smf.ols('log_footfalls~t', data=train).fit()
pred_exp = pd.Series(exp.predict(pd.DataFrame(test['t'])))
rmse_exp=np.sqrt(mean_squared_error(test['Footfalls'],pred_exp))
rmse_exp
```

Quadratic

```python
quad=smf.ols('Footfalls~t+t_square',data=train).fit()
pred_quad=pd.Series(quad.predict(pd.DataFrame(test[['t','t_square']])))
rmse_quad=np.sqrt(mean_squared_error(test['Footfalls'],pred_quad))
rmse_quad
```

```python
addi_sea =smf.ols('Footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train).fit()
pred_addi_sea=pd.Series(addi_sea.predict(pd.DataFrame(test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']])))
rmse_addi_sea=np.sqrt(mean_squared_error(test['Footfalls'],pred_addi_sea))
rmse_addi_sea
```

```python
add_sea =smf.ols('Footfalls~t+t_square+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train).fit()
pred_add_sea=pd.Series(add_sea.predict(pd.DataFrame(test[['t','t_square','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']])))
rmse_sea=np.sqrt(mean_squared_error(test['Footfalls'],pred_add_sea))
rmse_sea
```

multiplicative sea

```python
mul_sea =smf.ols('log_footfalls~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=train).fit()
pred_mul_sea=pd.Series(mul_sea.predict(pd.DataFrame(test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']])))
rmse_mul_sea=np.sqrt(mean_squared_error(test['Footfalls'],pred_mul_sea))
rmse_mul_sea
```

```python
#compare the results
data={'model':pd.Series(['rmse_linear','rmse_exp','rmse_quad','rmse_addi_sea','rmse_sea','rmse_mul_sea']),
      'RMSE_VAULE':pd.Series([rmse_linear,rmse_exp,rmse_quad,rmse_addi_sea,rmse_sea,rmse_mul_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_VAULE'])
```
