### A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
data=pd.read_csv('/content/Cutlets.csv')
```

```python
data.head()
```

We are going to use Two Tailed T test on independent sample with numeric data

```python
data.describe() #Descriptive Statistics
```

```python
data.info()#data type
```

Hypothesis using T Test

```python
test_statistic,P_value=stats.ttest_ind(data['Unit A'],data['Unit B'])
```

```python
print('P_value=',P_value)
```

Compare P_vaule with Significane vaule
if p vaule is not equal to signifacances vaule then we fail to reject Null Hypothesis(h0)

if p_vaule is equal to signifacances vaule then we reject null Hypothesis (h0)

```python
alpha = 0.05 # signifance vaule
```

```python
if  P_value < alpha :
  print('we reject the null hypothesis there is no significance difference between two Units A and B')
else:
  print('we fail to reject Null hypothesis')
```

**Hence, We fail to reject Null Hypothesis because of lack of evidence, there is  significant difference between the two samples**

#    A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.
### # 

```python
# importinng important lib
```

```python
import pandas as pd
import numpy as np 
```

```python
labtat=pd.read_csv("/content/LabTAT.csv")
```

```python
labtat.head()
```

```python
labtat.describe() # descriptive stat
```

```python
labtat.info()#data type 
```

Due to 4 independnt sample we are going to use ANOVA test

```python
_,p_value = stats.f_oneway(labtat.iloc[:,0],labtat.iloc[:,1],labtat.iloc[:,2],labtat.iloc[:,3])
print('p_value =',p_value)
```

```python
alpha = 0.05 # signifance vaule
```

```python
if  p_value < alpha :
  print('We reject Null Hypothesis there is a significance difference between TAT of reports of the laboratories')
else:
  print('we fail to reject Null hypothesis')
```

**Hence, We fail to reject Null Hypothesis because of lack of evidence, there is no significant difference between the two samples**

```python
observed = np.array([50, 142, 131, 70, 435, 1523, 1356, 750])
expected = np.array([42.76531299,  146.81287862,  131.11756787, 72.30424052, 442.23468701, 1518.18712138, 1355.88243213, 747.69575948])
```

```python
statistics, p_value = stats.chisquare(observed, expected, ddof = 3) #Comparing Evidence with Hypothesis
print("Statistics = ",statistics,'P_Value = ', p_value)
```

```python
alpha = 0.05 # significance vaule 
if p_value <= alpha:
  print('We reject Null Hypothesis there is a significance difference ')
else:
  print('We fail to reject Null hypothesis')
```

We fail to reject Null Hypothesis because of lack evidence. Therefore, there is no association or dependency between male-female buyers rations and are similar across regions.
